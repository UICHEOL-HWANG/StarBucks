```python
df['Beverage'].unique()
```

```python
df['Beverage'] = df.apply(lambda x: x['Beverage'].replace("Caff챔 Latte", "Cafe Latte"), axis=1)
```

```python
df['Beverage'] = df.apply(lambda x: x['Beverage'].replace("Shaken Iced Tazo짰 Tea (With Classic Syrup)", 
                                                          "Shaken Iced Tazo Tea (With Classic Syrup)"), axis=1)
```

```python
df['Beverage'] = df.apply(lambda x: x['Beverage'].replace("Shaken Iced Tazo짰 Tea Lemonade (With Classic Syrup)", 
                                                          "Shaken Iced Tazo Tea Lemonade (With Classic Syrup)"), axis=1)
```

```python
df['Beverage'] = df.apply(lambda x: x['Beverage'].replace("Java Chip', 'Strawberries & Cr챔me (Without Whipped Cream)", 
                                                          "Java Chip', 'Strawberries & Creame (Without Whipped Cream)"), axis=1)
```

```python
df['Beverage_category'].unique()
```

```python
df['Beverage_category'].nunique()
```

```python
df['Beverage_category'] = df.apply(lambda x: x['Beverage_category'].replace("Frappuccino짰", "Frappuccino"), axis=1)
```

```python
df['Beverage_category'] = df.apply(lambda x: x['Beverage_category'].replace("Frappuccino Blended Cr챔me", "Frappuccino Blended Creme"), axis=1)
```

```python
df['Beverage_category'] = df.apply(lambda x: x['Beverage_category'].replace("Tazo짰", "TazoWKT"), axis=1)
```

```python
plt.figure(figsize=(12,8),dpi=200)
sns.countplot(y=df['Beverage_category'], order=df['Beverage_category'].value_counts().index)
plt.xlabel("Count of Beverage categories")
plt.ylabel("Beverage Category")
plt.title("Beverage Category Vs. Count of Beverage Category in the DataFrame");
```

```python
plt.figure(figsize=(12,8),dpi=200)
sns.countplot(y=df['Beverage_prep'], order=df['Beverage_prep'].value_counts().index)
plt.xlabel("Count of Beverage Prep")
plt.ylabel("Beverage Prep")
plt.title("Beverage Prep Vs. Count of Beverage Prep in the DataFrame");
#음료재료중 비중이 가장 높은 순위
```

```python
df.groupby(['Beverage_category','Beverage', 'Beverage_prep']).max()['Calories'].sort_values(ascending=False)[:10] #음료카테고리,음료,음료재료들을 칼로리 단위로 그룹화 하여 정렬시킴 
```

```python
# Top 10 Drinks with High value of Calcium

df.groupby(['Beverage_category','Beverage', 'Beverage_prep']).max()['Calories'].sort_values(ascending=False)[:10].plot(kind='barh', color='#54B435')
plt.xlabel("Calories in the Drink")
plt.ylabel("Drink")
plt.title("Top 10 Drinks with High value of Calcium"); #칼로리가 가장 높은 음료  탑 10  
```
